# Clustering

**Clustering** (ou agrupamento) é uma técnica de aprendizado não supervisionado usada para identificar padrões ocultos e segmentar observações em grupos similares, sem utilizar rótulos prévios. O objetivo deste projeto é aplicar clustering para segmentação de clientes do iFood, com base em suas características demográficas e comportamentais — como idade, renda, consumo por categoria, número de filhos, entre outros.

Utilizamos técnicas de agrupamento não supervisionado para identificar perfis distintos de clientes que podem embasar ações de marketing direcionadas, como campanhas específicas, ofertas e fidelização.

Com base nos dados, buscamos responder: Que tipos de clientes compram mais? Quem engaja menos? Quem responde melhor às campanhas?

Em vez de rotular os clientes manualmente, o algoritmo identifica grupos naturais nos dados.

**A Importância do Pré-processamento**
O sucesso da clusterização depende da qualidade dos dados fornecidos ao algoritmo. Por isso, realizamos um pré-processamento cuidadoso, que inclui:

Limpeza de dados: remoção de inconsistências, valores faltantes e duplicações

Tratamento de variáveis categóricas: uso de OneHotEncoder para converter variáveis como Education e Marital_Status em formato numérico

Padronização (StandardScaler): essencial para algoritmos baseados em distância como o K-Means, evitando que variáveis com escalas maiores (como Income) dominem o processo.

Redução de dimensionalidade (PCA): melhora a visualização e pode reduzir ruído de variáveis pouco informativas

```python
one_hot_cols = [
    "Education",
    "Marital_Status",
    "AgeGroup",
    "HasChildren",
    "HasAcceptedCmp",
    "AcceptedCmp1",
    "AcceptedCmp2",
    "AcceptedCmp3",
    "AcceptedCmp4",
    "AcceptedCmp5",
    "Complain",
    "Response",
    "Kidhome",
    "Teenhome",
    "YearsSinceEnrolled",
    "Children",
    "AcceptedCmpTotal",
]

standard_cols = [
    "Income",
    "Age",
]

power_cols = df.columns[
   df.columns.str.startswith("Mnt") | df.columns.str.startswith("Num")
].tolist()

minmax_cols = [
    col for col in df.columns if col not in one_hot_cols + standard_cols + power_cols
]
```

```python
from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler,
    MinMaxScaler,
    PowerTransformer,
)
from sklearn.compose import ColumnTransformer

preprocessing = ColumnTransformer(
    [
        ("one-hot", OneHotEncoder(), one_hot_cols),
        ("standard", StandardScaler(), standard_cols),
        ("minmax", MinMaxScaler(), minmax_cols),
        ("power", PowerTransformer(), power_cols),
    ]
)

preprocessing.fit_transform(df)
```

```python
preprocessing.get_feature_names_out()
```

```python
df_scaled = pd.DataFrame(
    preprocessing.fit_transform(df),
    columns=preprocessing.get_feature_names_out()
)

df_scaled.head()
```

```python
scaled_columns_not_one_hot = df_scaled.columns[~df_scaled.columns.str.startswith("one")]
scaled_columns_not_one_hot
```

```python
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans

RANDOM_STATE = 42

pipeline = Pipeline(
    [
        ("preprocessing", preprocessing),
        ("clustering", KMeans(n_clusters=3, random_state=RANDOM_STATE, n_init=10))
    ]
)

pipeline.fit(df)
```

```python
pipeline.get_feature_names_out()
```

```python
pipeline.named_steps
```

```python
pipeline["clustering"].labels_
```

```python
df_clustered = df.copy()
df_clustered["cluster"] = pipeline["clustering"].labels_

df_clustered.head()
```
